```python
"""
Basic Command Line Calculator App
"""

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    return x / y

while True:
    print("Options:")
    print("Enter 'add' to add two numbers")
    print("Enter 'subtract' to subtract two numbers")
    print("Enter 'multiply' to multiply two numbers")
    print("Enter 'divide' to divide two numbers")
    print("Enter 'quit' to end the program")
    
    user_input = input(": ")

    if user_input == "quit":
        break
    
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))

    if user_input == "add":
        print("The answer is:", add(num1, num2))
    elif user_input == "subtract":
        print("The answer is:", subtract(num1, num2))
    elif user_input == "multiply":
        print("The answer is:", multiply(num1, num2))
    elif user_input == "divide":
        print("The answer is:", divide(num1, num2))
    else:
        print("Invalid input")
```

```python
"""
define x and y variables and assign variable ranging (1, 10) for x and an exponential function for y and plot x,y
"""
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 10, 100)
y = np.exp(x)

plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Exponential Function')
plt.show()
```

```python
repl_chain.run("create a dataframe using the above x and y values, x = rows, y = columns")
```

```python
"""
create a dataframe using the above x and y values, x = rows, y = columns
"""
import pandas as pd

data = {'x': x, 'y': y}
df = pd.DataFrame(data)

df
```

```python
"""
Scatter plot of x and y from the dataframe
"""

import matplotlib.pyplot as plt

plt.scatter(data['x'], data['y'])
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter plot of x and y")
plt.show()
```

```python
"""
Rename the dataframe to df_xy and plot the scatter plot again
"""
df_xy = df.copy()
plt.scatter(df_xy['x'], df_xy['y'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot of x and y')
plt.show()
```

```python
repl_chain.run("create a new dummy column named y_pred in the dataframe and fill it with values that follows a different exponential function from the current y")
```

```python
"""
Create a new dummy column named y_pred with values following a different exponential function from the current y
"""

def exponential_function(x, a=2, b=1.5):
    return a * (b ** x)

df_xy['y_pred'] = df_xy['x'].apply(exponential_function) # it produced dataframe instead of df_xy, have to include memory feature
```

```python
repl_chain.run("generate a regression report using x, y and y_pred from df_xy dataframe")
```

```python
"""
Generate a regression report using x, y and y_pred from df_xy dataframe
"""
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

x = df_xy['x']
y = df_xy['y']
y_pred = df_xy['y_pred']

mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)

regression_report = {
    'Mean Squared Error': mse,
    'Root Mean Squared Error': rmse,
    'Mean Absolute Error': mae,
    'R2 Score': r2
}

for key, value in regression_report.items():
    print(f'{key}: {value}')
```
